```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
data = pd.read_csv('data/USvideos.csv')
data.head()
```

```python
def parse_custom_date(date_str):
    try:
        parts = date_str.split('.')
        if len(parts) == 3:
            year = parts[0]
            day = parts[1]
            month = parts[2]
            return f'{day}-{month}-{year}'
    except ValueError:
        return pd.NaT
```

```python
data['publish_time'] = pd.to_datetime(data['publish_time'], format='mixed').dt.date
data
```

```python
data['trending_date'] = pd.to_datetime(data['trending_date'], format='ISO8601').dt.date
data
```

# Найкращі місяці по лайкам/переглядам

```python
data['publish_time'] = pd.to_datetime(data['publish_time'])
data['publish_month'] = data['publish_time'].dt.month
data['publish_month']
```

```python
df_month = data.groupby(['publish_month'])[['likes', 'views']].sum()
df_month = df_month.reset_index()
df_month
```

```python
data['publish_time'].max(), data['publish_time'].min()
```

```python
data['publish_year'] = data['publish_time'].dt.year
data['publish_year'].value_counts()
```

```python
plt.hist(df_month['likes'])
```

```python
sns.barplot(x=df_month['publish_month'], y=df_month['likes'], data=df_month)
plt.xlabel('Publish Month')
plt.ylabel('Total Likes')
plt.title('Total Likes per Month')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
